```python
def replace_low_frequency_with_mode(df):
    object_columns = df.select_dtypes(include='object').columns
    
    for column in object_columns:
        value_counts = df[column].value_counts()
        
        low_freq_values = value_counts[value_counts <= 100].index
        
        mode_value = df[column].mode()[0]
        
        df[column] = df[column].apply(lambda x: mode_value if x in low_freq_values else x)

    return df

df_train = replace_low_frequency_with_mode(df_train)
df_test = replace_low_frequency_with_mode(df_test)
```

# Preprocessing numerical column

```python
def detect_outliers_iqr(df):
    outliers = pd.DataFrame()
    for column in df.select_dtypes(include=['float64','int64']):
        Q1 = df[column].quantile(0.01)
        Q3 = df[column].quantile(0.99)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = ((df[column] < lower_bound) |(df[column] > upper_bound))
    return outliers
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_boxplot(df, title):
    plt.figure(figsize=(15,8))
    sns.boxplot(data=df.select_dtypes(include=['float64','int64']))
    plt.title(f'Boxplot of {title} Data',fontsize=16)
    plt.xticks(rotation=360, fontsize=12)
    plt.xlabel('Features',fontsize=14)
    plt.ylabel('Values',fontsize=14)
    plt.show()
```

```python
plot_boxplot(df_train, 'Train')
```

```python
def replace_outliers_with_median(df):
    df_copy = df.copy()
    for column in df.select_dtypes(include=['float64','int64']):
        Q1 = df[column].quantile(0.01)
        Q3 = df[column].quantile(0.99)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        median = df[column].median()
        df_copy[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)
    return df_copy
    
```

```python
train_cleaned = replace_outliers_with_median(df_train)
plot_boxplot(train_cleaned , 'Train(outliers replaced with median)')

test_cleaned = replace_outliers_with_median(df_test)
plot_boxplot(test_cleaned, 'Test(outliers replaced with median)')
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X_TRAIN_category = X_TRAIN.select_dtypes('object')
X_VAL_category = X_VAL.select_dtypes('object')
X_TEST_category = x_test.select_dtypes('object')
```

```python
enc = OneHotEncoder(handle_unknown = 'ignore',sparse_output = False).fit(X_TRAIN_category)
```

```python
X_TRAIN_OH = enc.transform(X_TRAIN_category)
X_VAL_OH = enc.transform(X_VAL_category)
X_TEST_OH = enc.transform(X_TEST_category)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_TRAIN_conti = X_TRAIN.select_dtypes('float64')
X_VAL_conti = X_VAL.select_dtypes('float64')
X_TEST_conti = x_test.select_dtypes('float64')
```

```python
scale = StandardScaler().fit(X_TRAIN_conti)
X_TRAIN_STS = scale.transform(X_TRAIN_conti)
X_VAL_STS = scale.transform(X_VAL_conti)
X_TEST_STS = scale.transform(X_TEST_conti)
```

# Preparing the input dataset

```python
X_TRAIN = np.concatenate([X_TRAIN_OH,X_TRAIN_STS],axis=1)
X_VAL = np.concatenate([X_VAL_OH,X_VAL_STS],axis=1)
X_TEST = np.concatenate([X_TEST_OH,X_TEST_STS],axis=1)
```

```python
Y_TRAIN = Y_TRAIN.values.ravel()
Y_VAL = Y_VAL.values.ravel()
```

```python
X_TRAIN.shape, X_VAL.shape, X_TEST.shape, Y_TRAIN.shape, Y_VAL.shape
```

# Model Training

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf_model = rf.fit(X_TRAIN, Y_TRAIN)
```

# Model Evaluation

```python
from sklearn.metrics import matthews_corrcoef
Y_VAL_pred = rf_model.predict(X_VAL)
mcc = matthews_corrcoef(Y_VAL , Y_VAL_pred)
print(f"Matthews correlation coefficient (MCC) on validation set: {mcc:.4f} ")
Y_TEST_pred = rf.predict(X_TEST)
```
